**Step 2:** Data Cleaning.

```python
# --- Step 2: Data Cleaning ---

# 1. Handle missing values for Age: Fill with the median
# Median is used because it is less sensitive to outliers (like very old passengers)
df['Age'] = df['Age'].fillna(df['Age'].median())

# 2. Handle missing values for Embarked: Fill with the mode
# Since Embarked is categorical, we use [0] to get the most frequent city ('S')
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# 3. Handle missing values for Cabin: Drop the column
# There are too many missing values in Cabin to fill them accurately
df.drop(columns=['Cabin'], inplace=True)

# 4. Remove duplicates
# This ensures every row in your dataset is unique
df.drop_duplicates(inplace=True)

# --- Verification ---
# Let's check if there are any null values left
print("Missing values per column:")
print(df.isnull().sum())

print("\nCleaning Complete! Here are the first 5 rows of your clean data:")
df.head()
```

**Step 3:** Data Analysis

```python
# --- Step 3: Data Analysis ---

# 1. Survival rate by Gender
# (Shows if being male or female changed your chances)
print("Survival rate by Gender:")
print(df.groupby('Sex')['Survived'].mean())

# 2. Survival rate by Class (Pclass)
# (1st class, 2nd class, or 3rd class)
print("\nSurvival rate by Class:")
print(df.groupby('Pclass')['Survived'].mean())

# 3. Average age per Class
print("\nAverage age per Class:")
print(df.groupby('Pclass')['Age'].mean())

# 4. Survival rate by Age Group
# First, we create age groups: Child (0-12), Teen (13-18), Adult (19-60), Senior (60+)
bins = [0, 12, 18, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

print("\nSurvival rate by Age Group:")
print(df.groupby('AgeGroup', observed=False)['Survived'].mean())
```

**Step 4:** Filtering

```python
# --- Step 4: Filtering ---

# 1. Female passengers who survived
# Uses & (AND) to ensure both conditions are true
females_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]

# 2. Children who survived
# Using the standard definition of children (under 18)
children_survived = df[(df['Age'] < 18) & (df['Survived'] == 1)]

# 3. Passengers in 1st class with high survival probability
# We filter for 1st class; later we can see their specific survival rate
first_class_passengers = df[df['Pclass'] == 1]

# --- Displaying the Filtered Results ---
print(f"✅ Total Female Survivors: {len(females_survived)}")
print(f"✅ Total Child Survivors: {len(children_survived)}")
print(f"✅ Total Passengers in 1st Class: {len(first_class_passengers)}")

# Preview one of the filters (e.g., the first 5 children who survived)
print("\nPreview of Children Survivors:")
children_survived.head()
```
